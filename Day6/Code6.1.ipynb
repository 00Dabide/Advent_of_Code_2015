```python
import requests
```

```python
data = requests.get("https://raw.githubusercontent.com/00Dabide/Advent_of_Code_2015/refs/heads/main/Day6/Day6Text.txt").text.split("\n")
```

```python
#make the 1k*1k grid
def make_grid():
    grid = []
    line = []
    for _ in range(0,1000):
        line.append(0)
    
    for _ in range(0,1000):
        grid.append(line)
    
    return grid

grid = make_grid()

#first value is y (vertical), second value is x (horizontal)
```

```python
# make instructions clear, so they can be readable

def get_instructions(data):
    pretty_instructions = []

    for instruction in data:
        sliced_instruction = instruction.split(" ")
        if len(sliced_instruction) == 5:
            clean_instruction = [sliced_instruction[0] + " " + sliced_instruction[1], sliced_instruction[2], sliced_instruction[4]]
        else:
            clean_instruction = [sliced_instruction[0], sliced_instruction[1], sliced_instruction[3]]
        pretty_instructions.append(clean_instruction)
    
    return pretty_instructions

clean_instructions = get_instructions(data)
        
```

```python
def get_lit_lights(instructions, grid):
    for instruction in instructions:

        y_start = int(instruction[1].split(",")[1])
        x_start = int(instruction[1].split(",")[0])

        

        y_end = int(instruction[2].split(",")[1])
        x_end = int(instruction[2].split(",")[0])

        for y_axis in range(y_start, y_end + 1):
            print(y_axis)
            for x_axis in range(x_start, x_end + 1):
            

                if instruction[0] == "turn on":
                    if grid[y_axis][x_axis] == 0:
                        grid[y_axis][x_axis] = 1
                        print(y_axis)


                elif instruction[0] == "turn off":
                    if grid[y_axis][x_axis] == 1:
                        grid[y_axis][x_axis] = 0


                else:
                    if grid[y_axis][x_axis] == 1:
                        grid[y_axis][x_axis] = 0
                    elif grid[y_axis][x_axis] == 0:
                        grid[y_axis][x_axis] = 1
        return 0
```
